Note: Review how to define a gloabl variable inside ```.query()```
 

````python
# Write your answer here:

features = pd.read_csv("data/features.csv")

threshold = 3000

weight_threshold = features.query("weight > @threshold")
weight_3000 = features.query("weight <= 3000")

mean_weight_threshold = np.mean(weight_threshold["acceleration"])
mean_weight_3000 = np.mean(weight_3000["acceleration"])

print(mean_weight_threshold)
print(mean_weight_3000)

print("The dataset with a weight greater than threshold has a mean acceleration of " + str(mean_weight_threshold) + ".")
print("The dataset with a weight less than or equal to threshold has a mean acceleration of " + str(mean_weight_3000) + ".")
````

(c) Regression

<font size = "3">

In this exercise you are asked to plot acceleration vs. weight and to plot <br>
 the best fit line.

- Run a linear regression between acceleration and weight.
- Extract the intercept and coefficient.
- Using your predicted coefficients, compute the predicted value of acceleration
- Plot a scatter plot with acceleration on the y axis and weight on the x axis.
- On that same figure, plot a line with the predicted values of acceleration

Do not forget to label your axes and title.

````python
# Write your answer here:

y = features["acceleration"]
x = features["weight"]

linreg = smf.ols(formula = 'y ~  x', data = features)
linreg_results = linreg.fit()

linreg_list = linreg_results.params
print(linreg_list)

b0 = 19.606158
b1 = -0.001359

features["p"] = b0 + b1*x

plt.scatter(x = x,y = y)
plt.plot(x, features["p"], color = 'green')

plt.xlabel("Weight")
plt.ylabel("Acceleration")
plt.title("Weight vs. Acceleration")
plt.legend(labels = ["Data points","Estimated Predicted Model"])
plt.show()
````
